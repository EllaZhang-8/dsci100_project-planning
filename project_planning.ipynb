## 4. Methods and plan

### Method: Linear regression

#### - Why is this method appropriate?

- In our dataset, the data is inbalanced for different age and experience level and skewed. linear regression is less sensitive to the inbalanced distribution of played hours. knn regression model depends more on distance to nearby points, so when many players have played hours 0, knn predictions will go toward zero.
- Although our data doesn't seem to be very linear, linear regression also allow us to test whether age or experience level has any effect on played hours, as our goal is to explain what kind of players can contribute more data.
- Linear regression gives coefficients (slope and vertical intercept) that can show how much played hours change when age or experience changes. knn model cannot provide this kind of explanation that it can only give predictions.

#### - Required assumptions

- The relationship between the predictors (age, experience) and played hours is roughly linear.
- The data not be clustered or repeated.
- The predictors should not be too strongly correlated with each other.

#### - Potential limitations or weaknesses

- If the true relationship is curved or nonlinear,  a linear model may give poor predictions and misleading coefficients.
- A few players have unusually high playtime (150–250 hours). These points can strongly mislead the regression line.

#### - How will I select and compare model?

1. Fit an initial linear regression model
2. Evaluate the model using linear coefficiencts
3. Use 5 fold cross-validation on the training dataset to chech how the model performs
4. Compare all model versions
5. Choose final model

#### -How to process data to apply the model?

1. Load and clean the data (before splitting)

3. Split the dataset into 80% for training and 20% for testing
